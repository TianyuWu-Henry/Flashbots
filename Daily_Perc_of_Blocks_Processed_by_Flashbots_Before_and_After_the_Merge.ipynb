# Before Merge

```python
onchain_block_before_merge = pd.read_csv('/content/drive/My Drive/WIP/block_data_before_merge.csv',index_col=0).sort_values("number").reset_index(drop=True)
onchain_block_before_merge
```

```python
onchain_block_before_merge_block_number = onchain_block_before_merge['number'].tolist()
len(onchain_block_before_merge_block_number)
```

```python
flashbot_block_before_merge = pd.read_csv('/content/drive/My Drive/WIP/flashbot_block_before_merge.csv',index_col=0).sort_values("block_number").reset_index(drop=True)
flashbot_block_before_merge
```

```python
flashbot_block_before_merge_block_number = flashbot_block_before_merge['block_number'].tolist()
len(flashbot_block_before_merge_block_number)
```

```python
result_before = []
for num in flashbot_block_before_merge_block_number:
    subtracted_num = int(num) - 15348255
    result_before.append(subtracted_num)
```

```python
daily = 7000
import numpy as np
period_before = int(np.floor(onchain_block_before_merge.shape[0] / daily))
period_before
```

```python
daily_perc_before_merge = []

range_start = 0
range_end_before = period_before * daily

# Define the range increment
range_increment = 7000

# Create a dictionary to store the appearance rates
appearance_rates_before = {}

# Count the appearance rates within each range
for i in range(range_start, range_end_before, range_increment):
    range_count_before = sum(i <= num < i + range_increment for num in result_before)
    appearance_rates_before[f"{int(i/7000 + 1)}"] = range_count_before / 7000
```

```python
df_appearance_rates_before_merge = pd.DataFrame(appearance_rates_before.items(), columns=['Day', 'Appearance Rate'])
df_appearance_rates_before_merge['status'] = 'before merge'
df_appearance_rates_before_merge
```

# After Merge

```python
onchain_block_after_merge = pd.read_csv('/content/drive/My Drive/WIP/block_data_after_merge.csv',index_col=0).sort_values("number").reset_index(drop=True)
onchain_block_after_merge
```

```python
onchain_block_after_merge_block_number = onchain_block_after_merge['number'].tolist()
len(onchain_block_after_merge_block_number)
```

```python
flashbot_block_after_merge = pd.read_csv('/content/drive/My Drive/WIP/flashbot_block_after_merge.csv',index_col=0).sort_values("block_number").reset_index(drop=True)
flashbot_block_after_merge
```

```python
flashbot_block_after_merge_block_number = flashbot_block_after_merge['block_number'].tolist()
len(flashbot_block_after_merge_block_number)
```

```python
result_after = []
for num in flashbot_block_after_merge_block_number:
    subtracted_num = int(num) - 15537940
    result_after.append(subtracted_num)
```

```python
period_after = int(np.floor(onchain_block_after_merge.shape[0] / daily))
```

```python
daily_perc_after_merge = []

range_start = 0
range_end_after = period_after * daily

# Define the range increment
range_increment = 7000

# Create a dictionary to store the appearance rates
appearance_rates_after = {}

# Count the appearance rates within each range
for i in range(range_start, range_end_after, range_increment):
    range_count = sum(i <= num < i + range_increment for num in result_after)
    appearance_rates_after[f"{int(i/7000 + 1)}"] = range_count / 7000
```

```python
df_appearance_rates_after_merge = pd.DataFrame(appearance_rates_after.items(), columns=['Day', 'Appearance Rate'])
df_appearance_rates_after_merge['status'] = 'after merge'
df_appearance_rates_after_merge
```

```python
df_appearance_rates = df_appearance_rates_before_merge.append(df_appearance_rates_after_merge).reset_index(drop=True)
```

```python
import plotly.express as px
fig1 = px.line(df_appearance_rates, y='Appearance Rate', color='status')
fig1.update_layout(
    title="Daily Perc of Blocks Processed by Flashbots Before and After Merge",
    xaxis_title='Day',
    yaxis_title='Percentage'
)
fig1.show()
```
